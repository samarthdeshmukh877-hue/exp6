```python
year = int(input("Enter a year: ")) #5

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
basic = float(input("Enter salary: ")) #7

if basic <= 10000:
   hra = basic*0.20
   da = basic*0.80
elif basic <= 20000:
   hra = basic*0.25
   da = basic*0.90
else:
   hra = basic*0.30
   da = basic*0.95

Gross = basic + hra + da

print("Gross salary:", Gross)
```

```python
annual_income = float(input("Enter your annual income: ")) #8

income_tax = 0.0
if annual_income <= 250000:
    income_tax = 0
elif annual_income <= 500000:
    income_tax = (annual_income - 250000) * 0.05
elif annual_income <= 1000000:
    income_tax = 12500 + (annual_income - 500000) * 0.20
else:
    income_tax = 12500 + 100000 + (annual_income - 1000000) * 0.30

print(f"Annual Income: {annual_income:,.2f}")
print(f"Calculated Income Tax: {income_tax:,.2f}")
```
